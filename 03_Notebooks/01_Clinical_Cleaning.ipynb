1. Load Clinical Excel

```python
import pandas as pd
import numpy as np

# Load file
clinical_path = "../01_Data_Raw/Tabular_Dataset/TOMPEI-CMMD_clinical_data_v01_20250121.xlsx"
df = pd.read_excel(clinical_path, sheet_name="Imaging Diagnosis Details Sheet")

print(df.columns)
df.head()
```

2. Edit Rows & Columns Manually

```python
# Removing Row 0 because it contains sub-headers for merged columns
df = df.drop(index=0).reset_index(drop=True)
# Renaming Columns Properly
df = df.rename(columns={
    "Unnamed: 4": "Exclusion reasons",
    "Mass": "Mass_Location",
    "Unnamed: 7": "Mass_Shape",
    "Unnamed: 8": "Mass_Margin",
    "Unnamed: 9": "Mass_Density",
    "Unnamed: 10": "Mass_Associated_calcification",
    "Unnamed: 11": "Mass_Other_Associated_Findings",
    "Calcification": "Calc_Location",
    "Unnamed: 13": "Calc_Morphology",
    "Unnamed: 14": "Calc_Distribution",
    "Unnamed: 15": "Calc_Clearly_Benign",
    "BI-RADS\nCategory": "BI_RADS"
})

print(df.columns)
df.head()
```

3. Remove Unnecessary Columns

```python
columns_to_keep = [
    "ID",
    "LeftRight",
    "Age",
    "classification",
    "Exclusion reasons",
    "Breast density",
    "BI_RADS",
    "Mass_Location",
    "Mass_Shape",
    "Mass_Margin",
    "Mass_Density",
    "Mass_Associated_calcification",
    "Mass_Other_Associated_Findings",
    "Calc_Location",
    "Calc_Morphology",
    "Calc_Distribution",
    "Calc_Clearly_Benign"
]

df = df[columns_to_keep].copy()

print(df.columns)
df.head()
```
