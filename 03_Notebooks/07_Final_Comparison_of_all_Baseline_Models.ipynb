```python
import matplotlib.pyplot as plt
import numpy as np

models = [
    'Baseline A (Random Forest)',
    'Baseline B (EfficientNet)',
    'Multimodal'
]

accuracy = [0.88, 0.89, 0.87]

macro_f1 = [0.78, 0.56, 0.62]

x = np.arange(len(models))

width = 0.35

plt.figure(figsize=(10,6))

plt.bar(
    x - width/2,
    accuracy,
    width,
    label='Accuracy'
)

plt.bar(
    x + width/2,
    macro_f1,
    width,
    label='Macro F1'
)

plt.xticks(x, models)

plt.ylim(0,1)

plt.ylabel("Score")

plt.title("Performance Comparison of All Baseline Models")

plt.legend()

plt.grid(axis='y')

plt.show()
```
